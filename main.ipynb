```python
from main import with_lstm, without_lstm, normal
import matplotlib.pyplot as plt 
import pandas as pd 
from src.utils import *
import os
%matplotlib inline
```

```python
arc_with_l, acc_with_lstm, time_with_lstm, history_with_lstm = with_lstm()
arc_without_l, acc_without_lstm, time_without_lstm, history_without_lstm = without_lstm()
arc_normal, acc_normal, time_normal, history_normal = normal()
```

```python
history_with_lstm_dict = convert_history_to_dict(history_with_lstm)
history_without_lstm_dict = convert_history_to_dict(history_without_lstm)
history_normal = history_normal.history
```

```python
plt.plot(history_with_lstm_dict['loss'], label='loss',ls='-.')
plt.plot(history_with_lstm_dict['accuracy'],label='accuracy',ls='-.')
plt.xlim(0,20)
plt.legend()
```

```python
plt.plot(history_without_lstm_dict['loss'], label='loss',ls='-.')
plt.plot(history_without_lstm_dict['accuracy'],label='accuracy',ls='-.')
plt.xlim(0,20)
plt.legend()
```

```python
plt.plot(history_normal['loss'], label='loss',ls='-.')
plt.plot(history_normal['accuracy'],label='accuracy',ls='-.')
plt.xlim(0,20)
plt.legend()
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history_with_lstm_dict['accuracy'], marker='D', label='With LSTM')
plt.plot(history_without_lstm_dict['accuracy'], marker='o', label='Without LSTM')
plt.plot(pd.DataFrame(history_normal)['accuracy'], marker='o', label='Normal')
plt.title('Accuracy Comparison')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.grid()
plt.xlim(0,17)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history_with_lstm_dict['loss'], marker='o', label='With LSTM')
plt.plot(history_without_lstm_dict['loss'], marker='o', label='Without LSTM')
plt.plot(pd.DataFrame(history_normal)['loss'], marker='o', label='Normal')
plt.title('Loss Comparison')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.grid()
plt.xlim(0, 17)
plt.legend()



plt.tight_layout()
plt.show()
```

```python
print("Time taken for model with LSTM:", time_with_lstm)
print("Time taken for model without LSTM:", time_without_lstm)
print("Time taken for normal model:", time_normal)
```
